```python

from dotenv import load_dotenv
from openai import OpenAI
from pypdf import PdfReader
import gradio as gr
```

```python
load_dotenv(override=True)
openai = OpenAI()
```

```python
def extract_pdf_text(pdf_file):
    """Reads uploaded PDF file and extracts all text."""
    reader = PdfReader(pdf_file)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text
```

```python
pdf_store = {"text": ""}
```

```python
def upload_pdf(pdf_file):
    if pdf_file is None:
        return "No PDF uploaded yet."
    pdf_text = extract_pdf_text(pdf_file.name)
    pdf_store["text"] = pdf_text
    return "✅ PDF uploaded and processed. You can now ask questions."
```

```python
def answer_question(question):
    """Answer question and clear the input box"""
    context = pdf_store["text"][:15000]  # Limit context for token safety
    prompt = f"""You are a helpful assistant. 
    Answer the question based on the following PDF content only.

    PDF Content:
    {context}

    Question: {question}
    Answer:"""

    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2,
    )
    
    # Return both the answer and clear the question box
    return response.choices[0].message.content.strip(), gr.update(value="")

    
    

```

```python
with gr.Blocks() as demo:
    gr.Markdown("## 📄 Maneesh's CoPilot")
    with gr.Row():
        pdf_input = gr.File(label="Upload a PDF", file_types=[".pdf"])
        upload_status = gr.Textbox(label="Status")
    upload_button = gr.Button("Process PDF")

    question_box = gr.Textbox(label="Ask a question about the PDF")
    answer_box = gr.Textbox(label="Answer")

    upload_button.click(fn=upload_pdf, inputs=pdf_input, outputs=upload_status)
    question_box.submit(fn=answer_question, inputs=question_box, outputs=[answer_box, question_box])
    
    
    
    

demo.launch(share=True)
```
